```python
plt.figure(figsize=(9,5))
ax = sns.countplot(x='Size',data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Amount'],kde=True)
plt.title('Distribution of Order Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

We can see distribution of the 'Amount'  to see how sales are distributed.



```python
plt.figure(figsize=(10, 6))
corr = df[['Qty', 'Amount']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
s_qty = df.groupby(['Size'],as_index=False).agg({"Qty":'sum'}).sort_values(by='Qty',ascending = False)
s_qty
```

```python
sns.barplot(x='Size',y='Qty',data=s_qty,palette='Set2')
plt.show()
```

Most of Qty brought is of M size in sales

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Courier Status',hue='Status',data=df,palette='Set2', linewidth=1.5, edgecolor='black')
plt.title("Order Status by Courier Status", fontsize=16, fontweight='bold')
plt.xlabel("Courier Status", fontsize=14)
plt.ylabel("Count of Orders", fontsize=14)
plt.legend(title="Order Status", title_fontsize='13', fontsize='12', loc='upper right')
plt.show()
```

Majority of orders 'shipped' through courier

```python
df['Size'].hist()
plt.show()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']

plt.figure(figsize=(12,6))

counts , bins , patches = plt.hist(column_data, bins = 10 ,color ='orange',linewidth =1.5, edgecolor = 'Black')

for patch , count in zip(patches , counts):
    height = patch.get_height()
    plt.text(patch.get_x() + patch.get_width()/2, height , int(count),
             ha = 'center',  va = 'bottom', fontsize = 13)


plt.title('Histogram of Category')
plt.xlabel('Category')
plt.ylabel('Frequency')

plt.xticks(rotation = 45)
plt.show()
```

Most of the buyers brought 'T-Shirts'

```python
b2b_check = df['B2B'].value_counts()
b2b_check 

plt.figure(figsize=(8,6))
plt.pie(b2b_check , labels = b2b_check.index, autopct = '%1.2f%%',
        colors=['#ff9999', '#66b3ff'], startangle=45, textprops={'fontsize': 12})


plt.legend(title="B2B Status", title_fontsize='13', fontsize='11', loc='upper right')
plt.title("Distribution of B2B Orders", fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

99.3% of buyers are 'retailers' & 0.7% are of b2b buyers

```python
x_data = df['Category']
y_data =df['Size']

plt.scatter(x_data , y_data,c='skyblue',edgecolor='black',alpha=0.7,marker='o')
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot of Category vs. Size', fontsize=16, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Demonstrates how many & which sizes are available across each category

```python
plt.figure(figsize=(12,6))
sns.countplot(x='ship-state',data = df)

plt.xlabel('Ship-State')
plt.ylabel('Count')
plt.title('Distriution by States',fontsize = 16,fontweight = 'bold')
plt.xticks(rotation = 90)
plt.show()
```
